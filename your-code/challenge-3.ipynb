First, I group the table by `CustomerID` getting only the sum of the amount spent.
Also, for Q2, I am adding a country column. Assuming that each client only lives/orders from one country.


```python
customers = orders.groupby('CustomerID').agg({'amount_spent':sum, 'Country':min})
customers
```

After, I assign those sums inside the quantiles:

```python
labels = ['Q1', 'Q2', 'Q3', 'Q4']
customers['Quantile'] = pd.qcut(customers['amount_spent'],4, labels=labels)
customers
```

Lastly, I apply a function to give customers in the fourth quantile the VIP status:

```python
def choose_VIP_status(x):
    if x == 'Q4':
        return True
    else:
        return False
    
def choose_preferred_status(x):
    if x == 'Q3':
        return True
    else:
        return False
```

```python
customers['VIP'] = customers['Quantile'].apply(choose_VIP_status)
customers['Preferred'] = customers['Quantile'].apply(choose_preferred_status)

customers
```

## Q2: How to identify which country has the most VIP Customers?

```python
# your code here
```

When grouping by customer id, I added their country, so now I only have to group by the country column and get the count of customers.

```python
countries = pd.DataFrame()
countries['Amount_of_customers'] = customers.groupby('Country')['VIP'].count()
countries['Amount_of_VIPs'] = customers.groupby('Country')['VIP'].sum()

countries.sort_values(by='Amount_of_VIPs', ascending=False)
```

## Q3: How to identify which country has the most VIP+Preferred Customers combined?

```python
# your code here
```

```python
countries['Amount_of_preferreds'] = customers.groupby('Country')['Preferred'].sum()
countries
```

```python
countries['Combined'] = countries['Amount_of_VIPs'] + countries['Amount_of_preferreds']
countries.sort_values(by='Combined', ascending=False)
```
